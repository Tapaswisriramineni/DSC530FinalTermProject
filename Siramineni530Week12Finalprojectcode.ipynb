## Identifying Outliers 

In this code, we first build the function identify_outliers_histogram, which accepts as arguments the dataset, the number of histogram bins (which are typically the square root of the number of rows), and the threshold for recognizing outliers.

The histogram of the data is calculated inside the function using np.histogram, and the mean and standard deviation are calculated using np.mean and np.std, respectively.

The threshold is a parameter that specifies how many standard deviations a number must deviate from the mean in order to be deemed an outlier. We define the outlier threshold as mean + threshold * std.


Using plt.hist, we plot the histogram, and plt.axvline, we add a vertical line at the outlier threshold. We use plt.scatter to plot any outliers above the threshold as green crosses.

## Hypothesis

My hypotheseis is to point out the no correlation between the Age and Airpollution variables which depicts that in the lung cancer prediction the Air pollution effect doesn't depend on the age of the person

```python
class permutatecorr(thinkstats2.HypothesisTest):
    """Tests correlations by permutation."""

    def TestStatistic(self, data):
        """Computes the test statistic.

        data: tuple of xs and ys
        """
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        """Run the model of the null hypothesis.

        returns: simulated data
        """
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
```

```python
# Creating a subset of dataset for Air and chronic_Lung_Disease
cleaned = lung_cancer_df.dropna(subset=['Air_Pollution', 'Age'])
cor_d = cleaned.Air_Pollution,cleaned.Age

# Running correlation test on data subset
cor_t = permutatecorr(cor_d)
pvalue = cor_t.PValue()

print ('pvalue: ', pvalue)
```

The pvalue shows that in the lung cancer prediction the air pollution effect is not dependent on the age of the patient

## Regression Analysis

```python
import statsmodels.formula.api as smf
```

```python
# Simple Linear Regression of Alcohol Use as a function of Genetic Risk

f = 'Alcohol_use ~ Genetic_Risk'
model = smf.ols(f, data=lung_cancer_df)
result = model.fit()
result.summary()
```

```python
# Extracting parameters
intercept = result.params['Intercept']
slope = result.params['Genetic_Risk']
print('Intercept:', intercept)
print('Slope:', slope)
```

```python
# Extracting p-value of slope estimate
s_p = result.pvalues['Genetic_Risk']
print('p-value of slope estimate:', s_p)
```

```python
# Extracting coefficent of determination
print('R^2:', result.rsquared)
```

An R-squared score of 0.7695 suggests that the independent variables in the model can account for around 76.95% of the variability in the dependent variable. This number ranges from 0 to 1, with 0 indicating that the model explains no variability and 1 indicating that it fully accounts for all variability.
